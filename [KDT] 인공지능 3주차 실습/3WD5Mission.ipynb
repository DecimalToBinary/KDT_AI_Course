- x-axis labelÍ≥º y-axis labelÏùÄ Ï†ÅÏ†àÌïú ÌÖçÏä§Ìä∏Î°ú Ï∂îÍ∞ÄÎêòÏñ¥Ïïº Ìï©ÎãàÎã§.

üí° Ïù¥ÎèôÌèâÍ∑†(Moving Average)Î≤ïÏùÄ ÏãúÍ≥ÑÏó¥ Îç∞Ïù¥ÌÑ∞Î•º ÌëúÌòÑÌïòÎäî Îç∞Ïóê ÎßéÏù¥ ÏÇ¨Ïö©ÌïòÎäî Î∞©Î≤ï Ï§ë ÌïòÎÇòÏûÖÎãàÎã§.

```python
BCDF=pd.read_csv("./BitCoin.csv")
BCDF
```

```python
BCDF["5-MA"]=BCDF["Open"].rolling(5).mean()
BCDF
```

```python
#"Date" ÏÜçÏÑ±ÏùÑ intÌòï Îç∞Ïù¥ÌÑ∞Î°ú Î≥ÄÌôò -> "Date"Í∞Ä 20160601~20170630 ÏÇ¨Ïù¥Ïù∏ Í∞úÏ≤¥ Ï∂îÏ∂úÌï¥ ÏÉà dataframeÏù∏ newBCDFÏóê Ï†ÄÏû•
newBCDF=BCDF[(pd.to_numeric(BCDF["Date"].str.replace("-",""))>=20160601) & (pd.to_numeric(BCDF["Date"].str.replace("-",""))<20170630)]
newBCDF
```

```python
SortedNewBCDF=newBCDF.sort_values(by="Date").reset_index(drop=True)
SortedNewBCDF
```

```python
len(SortedNewBCDF["Date"])
```

```python
plt.plot(SortedNewBCDF["Date"],SortedNewBCDF["5-MA"],color="#f2a900")
plt.xlabel("Date")
plt.ylabel("5 day moving average")

# "Date"Ïùò ÎààÍ∏à Ìè¨Îß∑ ÏÑ§Ï†ï
DFormat=mdates.DateFormatter('%Y-%m-%d')
#plt.subplot().xaxis.set_major_formatter(DFormat)
plt.axis(['2016-06-01','2017-06-30',0,5000])
plt.xticks([ i for i in range(0, len(SortedNewBCDF["Date"]), 30)])
plt.xticks(rotation=45)
plt.show()
#
```
